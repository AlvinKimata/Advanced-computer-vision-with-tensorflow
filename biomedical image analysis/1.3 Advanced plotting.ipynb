```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read in the 2D images.
im1 = imageio.imread('../inputs/TCIA Chest CT/chest-220.dcm')
im2 = imageio.imread('../inputs/TCIA Chest CT/chest-221.dcm')
im3 = imageio.imread('../inputs/TCIA Chest CT/chest-222.dcm')
```

```python
#Stack images into a volume.
vol = imageio.volread('../inputs/TCIA Chest CT/')
print('Volume dimensions:', vol.shape)
```

```python
#Init the figure and axes grid.
fig, axes = plt.subplots(nrows = 1, ncols = 2)

#Draw an image on each subplot.
axes[0].imshow(im1, cmap = 'gray')
axes[1].imshow(im2, cmap = 'gray')

#Remove the ticks/labels
axes[0].axis('off')
axes[1].axis('off')

plt.show()
```

## Slice 3D images.

### The simplest way to plot 3D and 4D images by slicing them into many 2D frames. Plotting many slices sequentially can create a "fly-through" effect that helps you understand the image as a whole.

```python
#Plot the images on a subplots array.
fig, axes = plt.subplots(nrows = 1, ncols = 4)

#Loop through subplots and draw image.
for ii in range(4):
    im = vol[ii, :, :]
    axes[ii].imshow(im, cmap = 'gray')
    axes[ii].axis('off')
    
#Render the image.
plt.show()
```

## Plot other views.

```python
vol.meta['sampling']
```

```python
#Select frame from vol.
im1 = vol[:, 256, :]
im2 = vol[:, :, 256]

#Compute the aspect ratio.
d0, d1, d2 = vol.meta['sampling']
asp1 = d0 / d2
asp2 = d0 / d1
```

```python
#Plot the images on a subplots array.
fig, axes = plt.subplots(nrows = 2, ncols = 1)

axes[0].imshow(im1, cmap = 'gray', aspect = asp1)
axes[1].imshow(im2, cmap = 'gray', aspect = asp2)

plt.axis('off')
plt.show()
```
