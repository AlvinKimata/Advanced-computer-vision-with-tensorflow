```python
import imageio
import matplotlib.pyplot as plt
```

```python
#Load the chest dcm image.
im = imageio.imread('../inputs/TCIA Chest CT/chest-221.dcm')
```

```python
#Print image attributes.
print('Image type:', type(im))
```

```python
print('Shape of the array:', im.shape)
```

```python
#Print the available metadata fields.
print(im.meta.keys())
```

```python
#Print modality data in the metadata.
print(im.meta['Modality'])
print(im.meta['sampling'])
print(im.meta['PatientSex'])
```

```python
#Plot images.
plt.imshow(im, cmap = 'gray', vmin = -200, vmax = 200)

#Render the image.
plt.show()
```

```python
#Remove the axis, ticks and labels.
plt.axis('off')
#Plot images.
plt.imshow(im, cmap = 'gray', vmin = -200, vmax = 200)
plt.show()
```
